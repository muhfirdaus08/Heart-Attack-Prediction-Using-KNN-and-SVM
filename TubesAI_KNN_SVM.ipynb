## Visualisasi dan Analisis Data

```python
df_all = df.copy()

df_more = df[df['output']==1]

df_less = df[df['output']==0]
```

```python
def plot_pie(column, title="All"):
    fig,axs = plt.subplots(1,1)
    data = df_all[column].value_counts()
    plt.pie(data,autopct='%1.2f%%',labels=data.index)
    plt.title(title)
    plt.show()
    
def plot_hist(column, title="all"):
    plt.hist(df_all[column],density=True)
    plt.title(title)
    plt.show()

def plot_bar(column, sort=False, title="all"):
    if sort:
        data_all = df_all[column].value_counts().sort_index()
    else:
        data_all = df_all[column].value_counts()
    plt.bar(data_all.index,data_all)
    plt.title(title)
    plt.show()
    
def plot_bar_compare(column, sort=False):
    if sort:
        data_churned = df_more[column].value_counts().sort_index()
        data_nonchurned = df_less[column].value_counts().sort_index()
    else:
        data_churned = df_more[column].value_counts()
        data_nonchurned = df_less[column].value_counts()
    
    fig,axs = plt.subplots(2,1)
    plt.subplots_adjust(left=0, bottom=0, right=1, top=2, wspace=0, hspace=0.2)
    axs[0].bar(data_nonchurned.index,data_nonchurned)
    axs[0].title.set_text('No Stroke')
    axs[1].bar(data_churned.index,data_churned)
    axs[1].title.set_text('Has Stroke')
    plt.show()

def plot_hist_compare(column, bins=5):
    plt.hist([df_less[column], df_more[column]] , color=['c','r'])
    plt.legend(('Less Chance', 'More Chance'))
    plt.show()
    
def plot_pie_compare(column):
    data_churned = df_more[column].value_counts()
    data_nonchurned = df_less[column].value_counts()
    
    fig,axs = plt.subplots(2,1)
    plt.subplots_adjust(left=0, bottom=0, right=1, top=2, wspace=0, hspace=0.2)
    axs[0].pie(data_nonchurned,autopct='%1.2f%%',labels=data_nonchurned.index)
    axs[0].title.set_text('Less Chance')
    axs[1].pie(data_churned,autopct='%1.2f%%',labels=data_churned.index)
    axs[1].title.set_text('More Chance')
    plt.show()

def plot_boxplot(column, title=""):
    sns.boxplot(x="output", y=column, palette=["c", "r"],
            hue="output",  data=df_all).set_title(title, fontsize=15)

def check_median(column):
    data_churned = df_more[column].describe()
    data_nonchurned = df_less[column].describe()
    print('Less Chance: {}'.format(data_nonchurned['50%']))
    print('More Chance: {}'.format(data_churned['50%']))

def check_most(column):
    data_churned = df_more[column].value_counts()
    data_nonchurned = df_less[column].value_counts()
    print('Less Chance: {}'.format(data_nonchurned.index[0]))
    print('More Chance: {}'.format(data_churned.index[0]))
```

## Seleksi Fitur

```python
cat_cols = ['sex','exng','caa','cp','slp','thall']
con_cols = ["age","thalachh","oldpeak"]

# encoding colum kategori
df = pd.get_dummies(df, columns = cat_cols, drop_first = True)

# menentukan colom fitur dan targer
x = df.drop(columns=['output',"chol","trtbps","fbs",'restecg'])
y = df['output']

# scaling fitur kontinue
x[con_cols] = scaler.fit_transform(x[con_cols])
```

## Split Data

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

## Metode KNN

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# mencari parameter terbaik
knn0 = KNeighborsClassifier()
knn_cv = GridSearchCV(knn0, {"n_neighbors": np.arange(1,50)}, cv=10)
knn_cv.fit(x,y)
print("Best parameters of KNN :",knn_cv.best_params_)
```

#### Training Data

```python
model_knn = KNeighborsClassifier(n_neighbors=knn_cv.best_params_["n_neighbors"])
model_knn.fit(X_train,y_train)
```

#### Testing Data

```python
y_predKNN = model_knn.predict(x_test)
y_predKNN
```

```python
y_test = np.array(y_test)
y_test 
```

#### Evaluasi Data

```python
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
acc_KNN = accuracy_score(y_test,y_predKNN)*100
acc_KNN
```

```python
prec_KNN = precision_score(y_test,y_predKNN)*100
prec_KNN
```

```python
rec_KNN = recall_score(y_test,y_predKNN)*100
rec_KNN
```

```python
f1_KNN = f1_score(y_test,y_predKNN)*100
f1_KNN
```

```python
from sklearn.metrics import confusion_matrix

knn_prediction = model_knn.predict(x_test)
plt.figure()
sns.heatmap(confusion_matrix(y_test, knn_prediction), annot=True, cmap='Blues')
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.title("KNN Confusion Matrix");
```

```python
from sklearn.metrics import classification_report
knn_report = classification_report(y_test, y_predKNN)
print(f'Classification Report KNN:\n{knn_report}')
```

## Metode SVM

```python
from sklearn.svm import SVC

# mencari parameter terbaik
grid = {"C":np.arange(1,10,1),'gamma':[0.00001,0.00005, 0.0001,0.0005,0.001,0.005,0.01,0.05,0.1,0.5,1,5]}
svm0 = SVC(random_state=42)
svm_cv = GridSearchCV(svm0, grid, cv=10)
svm_cv.fit(x,y)
print("Best parameters of SVC :",svm_cv.best_params_)
```

#### Training Data

```python
model_svm = SVC(C=svm_cv.best_params_["C"], gamma=svm_cv.best_params_["gamma"],random_state=42)
model_svm.fit(X_train,y_train)
```

#### Testing Data

```python
y_predSVM = model_svm.predict(x_test)
y_predSVM
```

```python
y_test = np.array(y_test)
y_test 
```

#### Evaluasi Data

```python
acc_SVM = accuracy_score(y_test,y_predSVM)*100
acc_SVM
```

```python
prec_SVM = precision_score(y_test,y_predSVM)*100
prec_SVM
```

```python
rec_SVM = recall_score(y_test,y_predSVM)*100
rec_SVM
```

```python
f1_SVM = f1_score(y_test,y_predSVM)*100
f1_SVM
```

```python
svm_prediction = model_svm.predict(x_test)
plt.figure()
sns.heatmap(confusion_matrix(y_test, svm_prediction), annot=True, cmap='Blues')
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.title("SVM Confusion Matrix");
```

```python
svm_report = classification_report(y_test, y_predSVM)
print(f'Classification Report SVM:\n{svm_report}')
```

Metode SVM memberikan akurasi yang lebih tinggi dibandingkan dengan metode KNN dengan memberikan parameter terbaik sesuai dengan data yang ada. Akurasi pada metode SVM sebesar 93% sedangkan metode KNN sebesar 88%
